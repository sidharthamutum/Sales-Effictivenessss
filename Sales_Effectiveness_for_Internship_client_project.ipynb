```python
# HANDLING MISSING VALUE

#Identify feature types
cat_feature = ['Source','Sales_Agent','Location','Delivery_Mode']
num_features = ['Product_ID']

# Fill categorical missing values
for col in cat_feature:
    df[col].fillna('Unknown', inplace=True)

# Fill numerical missing values with median
for col in num_features:
    df[col].fillna(df[col].median(), inplace=True)

# Verify missing values
print("Missing Values After Filling:\n", df.isnull().sum())
```

```python
# REMOVE DUPLICATE

# Remove duplicates
df.drop_duplicates(inplace=True)

# Check shape after removing duplicates
print("Shape After Removing Duplicates:", df.shape)

# Confirm no duplicates remain
print("Remaining Duplicates:", df.duplicated().sum())
```

```python
# DROPPED IRRELAVENT COLUMN

#Drop irrelevant columns that are not useful for prediction
# 'EMAIL' and 'Mobile' have likely been dropped already
irrelevant_cols = ['Created', 'year', 'Status_Encoded', 'status_encoded'] # Columns to drop that are still in df

#Drop the columns, using errors='ignore' in case some were already dropped
df.drop(columns=irrelevant_cols, inplace=True, errors='ignore')

#Verify remaining columns
print("Columns After Dropping Irrelevant Features:")
print(df.columns.tolist())

#Check current shape
print("Current Dataset Shape:", df.shape)
```

```python
# CONVERT CATEGORICAL FEATURE AND ENCODING

from sklearn.preprocessing import LabelEncoder, OneHotEncoder

# Encode target variable
le = LabelEncoder()
df['Status'] = le.fit_transform(df['Status'])  # High=1, Low=0

# Verify encoding
print("Unique Encoded Status Values:", df['Status'].unique())

# Identify Categorical Features to Encode
cat_features = ['Source', 'Sales_Agent', 'Location', 'Delivery_Mode']
print("Categorical Features to Encode:", cat_features)

# Apply One-Hot Encoding
f_encoded = pd.get_dummies(df, columns=cat_features, drop_first=True)

# Preview encoded data
print("Encoded Dataset Shape:", df_encoded.shape)
print(df_encoded.head())

# Verify Dataset After Encoding
# Check remaining data types
print("\nData Types After Encoding:")
print(df_encoded.dtypes.value_counts())

# Confirm no object columns remain
print("\nRemaining object columns:", df_encoded.select_dtypes('object').columns.tolist())
```

```python
# FEATURE ENGINEERING

# Convert 'Created' to datetime if it exists, handling errors
if 'Created' in df.columns:
    df['Created'] = pd.to_datetime(df['Created'], errors='coerce')

    # Extract date-based features
    df['Year'] = df['Created'].dt.year        # Capture yearly trend
    df['Month'] = df['Created'].dt.month      # Capture seasonal trend
    df['Weekday'] = df['Created'].dt.weekday  # 0=Monday, 6=Sunday

    # Create business-driven features
    # Lead age in days (handle cases where 'Created' is NaT due to coercion)
    df['Lead_Age_Days'] = (datetime.now() - df['Created']).dt.days
    df['Lead_Age_Days'] = df['Lead_Age_Days'].fillna(df['Lead_Age_Days'].median()) # Fill NaNs created by NaT in 'Created'

    # Weekend flag: 1 if Saturday/Sunday, else 0
    df['Is_Weekend'] = df['Weekday'].apply(lambda x: 1 if x in [5, 6] else 0)

    # Drop original 'Created' column
    df.drop('Created', axis=1, inplace=True)
else:
    print("'Created' column not found. Skipping date-based feature engineering.")
    # If 'Created' is already dropped, we need to handle 'month' column if it exists
    if 'month' in df.columns:
      df.rename(columns={'month': 'Month'}, inplace=True)
      # Create a dummy Year and Weekday if 'Created' is not available
      df['Year'] = 0 # Placeholder
      df['Weekday'] = 0 # Placeholder
      df['Lead_Age_Days'] = 0 # Placeholder
      df['Is_Weekend'] = 0 # Placeholder


# Handle missing values in 'Mobile' by filling with a placeholder
df['Mobile'].fillna('Unknown_Mobile', inplace=True)

# Clean '#VALUE!' in 'EMAIL' by replacing with a placeholder
df['EMAIL'].replace('#VALUE!', 'Unknown_Email', inplace=True)
df['EMAIL'].fillna('Unknown_Email', inplace=True) # Also fill any potential NaNs in EMAIL

# Combine rare categories in categorical features (<1% frequency)
categorical_cols_to_process = ['Source', 'Sales_Agent', 'Location'] # Delivery_Mode has only 5 unique values
for col in categorical_cols_to_process:
    if col in df.columns:
        freq = df[col].value_counts(normalize=True)
        rare_categories = freq[freq < 0.01].index
        df[col] = df[col].replace(rare_categories, 'Other')
    else:
        print(f"Column '{col}' not found. Skipping rare category combination for this column.")


# Identify Categorical Features to Encode (excluding 'Status' which is already encoded)
# Also exclude 'Mobile' and 'EMAIL' as they are less likely to be useful after simple imputation
cat_features_to_encode = ['Source', 'Sales_Agent', 'Location', 'Delivery_Mode']

# Apply One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=cat_features_to_encode, drop_first=True)

# Verify new features and encoding
print("Columns after feature engineering and encoding:")
print(df_encoded.columns.tolist())
print(df_encoded.head())

# Check for remaining missing values
print("\nMissing values after feature engineering and encoding:\n", df_encoded.isnull().sum().sum())
```

```python
# Initialize the model
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000 , random_state=42),
    'Decision Tree': DecisionTreeClassifier(random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=200, random_state=42),
    'XGBoost': XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
}

# Drop 'Mobile' and 'EMAIL' columns from x_train and x_test
x_train_cleaned = x_train.drop(['Mobile', 'EMAIL'], axis=1)
x_test_cleaned = x_test.drop(['Mobile', 'EMAIL'], axis=1)


# Train each model
for name, model in models.items():
    model.fit(x_train_cleaned, y_train)
    print(f"{name} trained.")

# Predict and evaluation
result = []

for name, model in models.items():
  y_pred = model.predict(x_test_cleaned)

  accuracy  = accuracy_score(y_test, y_pred)
  precision = precision_score(y_test, y_pred, average='weighted') # Use weighted average for multiclass
  recall    = recall_score(y_test, y_pred, average='weighted')    # Use weighted average for multiclass
  f1        = f1_score(y_test, y_pred, average='weighted')        # Use weighted average for multiclass


  result.append([name, accuracy, precision, recall, f1])

  print(f"\n{name} Evaluation:")
  print("confusion matrix:")
  print(confusion_matrix(y_test, y_pred))
  print("\nClassification Report:")
  print(classification_report(y_test, y_pred))
```

```python
# CREATE AND COMPARISION TABLE

# Collect Evaluation Metrics for Each Model
# Create dataframe
result_df = pd.DataFrame(result, columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1-Score'])

# Sort by F1-score
result_df = result_df.sort_values(by='F1-Score', ascending=False)

# Print the Comparison Table
# Show model comparision ()
print("\n model performance comparison:")
display(result_df)
```

HYPERPARAMETER TUNING

```python
# Importing the require libraries
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, classification_report
import pandas as pd
```

```python
# Define paragrid

# Logistic regression
param_grid_lr = {
    'penalty': ['l2', None],
    'C': [0.1, 1, 10],
    'solver': ['lbfgs', 'liblinear']
}

# Decision tree
param_grid_dt = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Random forest
param_grid_rf = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# XGBoost
param_grid_xgb = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.8, 0.9, 1.0],
    'colsample_bytree': [0.8, 0.9, 1.0]
}


# Create function for tuning and evaluation
def tune_and_evaluate(model, param_grid, model_name):
  Grid = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=1)
  Grid.fit(x_train_cleaned, y_train)

  best_model = Grid.best_estimator_
  y_pred = best_model.predict(x_test_cleaned)

  accuracy  = accuracy_score(y_test, y_pred)
  precision = precision_score(y_test, y_pred, average='weighted')
  recall    = recall_score(y_test, y_pred, average='weighted')
  f1        = f1_score(y_test, y_pred, average='weighted')

  print(f"\n {model_name} Best Params:", Grid.best_params_)
  print(f"📊 {model_name} Performance:")
  print(f"Accuracy: {accuracy:.4f}")
  print(f"Precision: {precision:.4f}")
  print(f"Recall: {recall:.4f}")
  print(f"F1-Score: {f1:.4f}")
  print("\nClassification Report:")
  print(classification_report(y_test, y_pred))

  return [model_name, accuracy, precision, recall, f1, best_model]

# Run Tuning for All Models
results = []

# Logistic Regression
results.append(tune_and_evaluate(LogisticRegression(max_iter=1000, random_state=42), param_grid_lr, "Logistic Regression"))

# Decision Tree
results.append(tune_and_evaluate(DecisionTreeClassifier(random_state=42), param_grid_dt, "Decision Tree"))

# Random Forest
results.append(tune_and_evaluate(RandomForestClassifier(random_state=42), param_grid_rf, "Random Forest"))

# XGBoost
results.append(tune_and_evaluate(XGBClassifier(eval_metric='logloss', random_state=42, use_label_encoder=False), param_grid_xgb, "XGBoost"))
```

```python
# Compare Tuned Models & Select Best
result_df = pd.DataFrame(results, columns=['Model', 'Accuracy', 'Precision', 'Recall', 'F1-Score', 'Best Model'])
result_df = result_df.sort_values(by='F1-Score', ascending=False)

print("\n Tuned Model Performance Comparison:")
display(result_df[['Model','Accuracy', 'Precision', 'Recall', 'F1-Score']])

# Select the best model
best_model_row = result_df.iloc[0]
best_model_name = best_model_row['Model']
best_model = best_model_row['Best Model']

print(f"\n best tuned model is: {best_model_name}")

```

FEATURE IMPORTANCE

```python
# Importing all the libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.inspection import permutation_importance
import shap


def feature_importance(model, x_train, x_test, y_test, model_name="", run_shap=False):
    """Combined feature importance Analysis : Native, Permutation, and optional SHAP.

    Parameters:
    model: Trained model
    X_train: Training features (DataFrame)
    X_test: Test features (DataFrame)
    y_test: Test target (Series)
    model_name: Optional name of the model (string)
    run_shap: Whether to run SHAP analysis (bool)
    """

    # Naive feature importance

    if hasattr(model, 'feature_importances_'):
        importances = model.feature_importances_
        fit_imp_df = pd.DataFrame({
            'Feature': x_train.columns,
            'Importance': importances
        }).sort_values(by='Importance', ascending=False)

        print("\nFeature Importance (Native):")
        print(fit_imp_df)

        plt.figure(figsize=(10, 6))
        plt.barh(fit_imp_df['Feature'][:20], fit_imp_df['Importance'][:20], color='skyblue') # Changed to barh and color
        plt.gca().invert_yaxis()
        plt.title(f'Feature Importance for {model_name} (Top 20)')
        plt.xlabel('Importance')
        plt.ylabel('Feature')
        plt.show()
    else:
        print("\n Naive feature importance is not available for this model.")

    # permutation importance
    print("\nPermutation Importance:")
    perm_imp = permutation_importance(
        model,
        x_test,
        y_test,
        n_repeats=30,
        random_state=42,
        n_jobs=-1
    )
    perm_imp_df = pd.DataFrame({
        'Feature': x_train.columns,
        'Importance': perm_imp.importances_mean
    }).sort_values(by='Importance', ascending=False)

    print("\nPermutation Importance:")
    print(perm_imp_df)

    plt.figure(figsize=(10, 6))
    plt.barh(perm_imp_df['Feature'][:20], perm_imp_df['Importance'][:20], color='salmon') # Changed to barh and color
    plt.gca().invert_yaxis()
    plt.title(f'Permutation Importance for {model_name} (Top 20)')
    plt.xlabel('mean Importance Decrease')
    plt.show()

    # Sharp Analysis
    if run_shap:
        print("\n running SHAP Analysis:")
        try:
            explainer = shap.TreeExplainer(model)
            shap_values = explainer.shap_values(x_train)

            shap.summary_plot(shap_values, x_train, feature_names=x_train.columns) # Changed feature_name to feature_names
        except Exception as e:
            print(f"SHAP analysis failed: {e}")


    print("\n feature importance analysis completed.")

# Example usage after traning:
# Feature_importance_analysis(best_model, X_train, X_test, y_train, y_test, model_name="XGboost", Run_shap=True)
```

# Challenges Faces


* Data Quality & Cleaning

Dataset contained missing values in both numerical and categorical columns.

Duplicate records were found and removed to avoid bias.

Some irrelevant columns (e.g., Email, Mobile) had to be dropped because they did not contribute to prediction.

*  Handling Categorical Variables

Several features were categorical with many unique levels.

Deciding between Label Encoding and One-Hot Encoding was important to balance model performance vs. dimensionality.

* Imbalanced Target Variable

The target column (Status) had unequal distribution between categories.

This made accuracy a misleading metric — required focusing on Precision, Recall, and F1-score.

* Feature Engineering

Extracting useful information from date columns (like Created) was tricky.

Needed to generate Month, Year, Weekday features and test their impact.

Some engineered features did not improve performance and were discarded.

* Outliers

Outliers were present in numerical features.

Challenge: deciding whether to remove them (risk losing important cases) or keep them (risk model distortion).

* Model Selection

Trade-off between simple and complex models:

Logistic Regression → easy to interpret but weaker performance.

Decision Tree → interpretable but prone to overfitting.

Random Forest/XGBoost → better accuracy but less explainable.

* Hyperparameter Tuning

Running GridSearchCV for Random Forest and XGBoost was computationally heavy.

Required optimizing parameters while managing training time.

* Model Evaluation

Needed to go beyond accuracy due to class imbalance.

Balancing Precision vs. Recall was challenging depending on business needs (avoiding false positives vs. false negatives).

* Interpretability

Stakeholders required clear insights from the model.

Black-box models (Random Forest, XGBoost) were harder to explain.

Used feature importance and suggested SHAP for transparency.

* Deployment Considerations

Ensuring preprocessing steps (encoding, scaling) are applied consistently on new/unseen data.

Saving the final trained model (joblib) and pipeline for reproducibility.

Planning for regular retraining as new sales data arrives.
